```python
# Google Colab ã§ n_queené–¢æ•°ã®ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆã‚’æãã‚³ãƒ¼ãƒ‰
# Colab ã§å®Ÿè¡Œã—ã¦ãã ã•ã„ï¼

# ========================================
# æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆè¨­å®šï¼ˆæ–‡å­—åŒ–ã‘å¯¾ç­–ï¼‰
# ========================================

print("ğŸ“¦ å¿…è¦ãªãƒ‘ãƒƒã‚±ãƒ¼ã‚¸ã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ä¸­...")

import subprocess
import sys
import os

# Graphvizã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
subprocess.run(['pip', 'install', '-q', 'graphviz'], check=False)

# æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
print("ğŸ“ æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ä¸­...")
subprocess.run(['apt-get', 'install', '-y', '-qq', 'fonts-noto-cjk'],
               stdout=subprocess.DEVNULL,
               stderr=subprocess.DEVNULL,
               check=False)

subprocess.run(['fc-cache', '-fv'],
               stdout=subprocess.DEVNULL,
               stderr=subprocess.DEVNULL,
               check=False)

print("âœ… ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«å®Œäº†!\n")

from graphviz import Digraph
from IPython.display import display
os.environ['LANG'] = 'ja_JP.UTF-8'

def create_n_queen_flowchart():
    """n_queené–¢æ•°ã®ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆã‚’ç”Ÿæˆã—ã¾ã™ï¼ˆå†å¸°çš„ãƒãƒƒã‚¯ãƒˆãƒ©ãƒƒã‚­ãƒ³ã‚°ï¼‰ã€‚"""

    dot = Digraph(comment='n_queen Function Flowchart')
    dot.attr(rankdir='TB')  # Top to Bottom æ–¹å‘
    # æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆè¨­å®šï¼ˆæ–‡å­—åŒ–ã‘å¯¾ç­–ï¼‰
    dot.attr('node', fontname='Noto Sans CJK JP, DejaVu Sans, sans-serif')
    dot.attr('edge', fontname='Noto Sans CJK JP, DejaVu Sans, sans-serif')
    dot.attr('graph', fontname='Noto Sans CJK JP, DejaVu Sans, sans-serif')

    # ãƒãƒ¼ãƒ‰å®šç¾©
    # é–‹å§‹
    dot.node('start', 'é–‹å§‹\\nn_queen(y)', shape='ellipse', style='filled', fillcolor='lightgreen')

    # å…¥åŠ›
    dot.node('input', 'å…¥åŠ›: y\\n(ç¾åœ¨ã®è¡Œç•ªå·)', shape='parallelogram', style='filled', fillcolor='lightyellow')

    # æ¡ä»¶1: y == N?
    dot.node('check_n', 'y == N ?\\n(å…¨è¡Œé…ç½®å®Œäº†?)', shape='diamond', style='filled', fillcolor='lightblue')

    # æˆåŠŸãƒ‘ã‚¹
    dot.node('success', 'é…ç½®æˆåŠŸï¼\\nprint_a()\\n\\nreturn 1\\n\\n[çµ‚äº†]',
             shape='box', style='filled,rounded', fillcolor='#ccffcc')

    # forãƒ«ãƒ¼ãƒ—é–‹å§‹
    dot.node('for_start', 'for x = 0 to N-1\\n(å„åˆ—ã‚’æ¢ç´¢)', shape='box', style='filled', fillcolor='#ffffcc')

    # q_checkå‘¼ã³å‡ºã—
    dot.node('q_check', 'q_check(x, y)\\n(é…ç½®å¯èƒ½ã‹ç¢ºèª)', shape='box', style='filled', fillcolor='#e6f3ff')

    # q_checkã®çµæœåˆ¤å®š
    dot.node('check_result', 'é…ç½®å¯èƒ½?', shape='diamond', style='filled', fillcolor='lightblue')

    # é…ç½®ä¸å¯ã®å ´åˆ
    dot.node('next_x', 'æ¬¡ã®xã¸', shape='box', style='filled', fillcolor='#ffe6e6')

    # ã‚¯ã‚¤ãƒ¼ãƒ³é…ç½®
    dot.node('place_queen', 'ã‚¯ã‚¤ãƒ¼ãƒ³ã‚’é…ç½®:\\nq[y] = x\\nflag_x[x] = 1\\nflag_n1[x+y] = 1\\nflag_n2[y-x+(N-1)] = 1',
             shape='box', style='filled', fillcolor='#e6ffe6')

    # å†å¸°å‘¼ã³å‡ºã—
    dot.node('recursive', 'n_queen(y+1)\\n(å†å¸°å‘¼ã³å‡ºã—)', shape='box', style='filled', fillcolor='#ffe6ff')

    # å†å¸°ã®çµæœåˆ¤å®š
    dot.node('recursive_result', 'å†å¸°å‘¼ã³å‡ºã—ã®\\nçµæœ == 1?', shape='diamond', style='filled', fillcolor='lightblue')

    # ã‚¯ã‚¤ãƒ¼ãƒ³å‰Šé™¤ (æˆåŠŸæ™‚)
    dot.node('remove_success', 'ã‚¯ã‚¤ãƒ¼ãƒ³ã‚’å‰Šé™¤:\\nq[y] = -1\\nflag_x[x] = 0\\nflag_n1[x+y] = 0\\nflag_n2[y-x+(N-1)] = 0\\n\\næ¬¡ã®è§£ã‚’æ¢ç´¢',
             shape='box', style='filled', fillcolor='#fff0e6')

    # ã‚¯ã‚¤ãƒ¼ãƒ³å‰Šé™¤ (å¤±æ•—æ™‚)
    dot.node('remove_fail', 'ã‚¯ã‚¤ãƒ¼ãƒ³ã‚’å‰Šé™¤:\\nq[y] = -1\\nflag_x[x] = 0\\nflag_n1[x+y] = 0\\nflag_n2[y-x+(N-1)] = 0',
             shape='box', style='filled', fillcolor='#ffe6e6')

    # forãƒ«ãƒ¼ãƒ—çµ‚äº†ç¢ºèª
    dot.node('for_end', 'å…¨ã¦ã®xã‚’\\næ¢ç´¢æ¸ˆã¿?', shape='diamond', style='filled', fillcolor='lightblue')

    # å¤±æ•—çµ‚äº†
    dot.node('fail', 'é…ç½®å¤±æ•—\\n\\nreturn 0\\n\\n[çµ‚äº†]',
             shape='box', style='filled,rounded', fillcolor='#ffcccc')

    # ã‚¨ãƒƒã‚¸(çŸ¢å°)å®šç¾©
    dot.edge('start', 'input')
    dot.edge('input', 'check_n')
    dot.edge('check_n', 'success', label='Yes')
    dot.edge('check_n', 'for_start', label='No')

    dot.edge('for_start', 'q_check')
    dot.edge('q_check', 'check_result')
    dot.edge('check_result', 'next_x', label='No')
    dot.edge('check_result', 'place_queen', label='Yes')

    dot.edge('place_queen', 'recursive')
    dot.edge('recursive', 'recursive_result')

    dot.edge('recursive_result', 'remove_success', label='Yes\\n(æˆåŠŸ)')
    dot.edge('recursive_result', 'remove_fail', label='No\\n(å¤±æ•—)')

    dot.edge('remove_success', 'for_end')
    dot.edge('remove_fail', 'for_end')
    dot.edge('next_x', 'for_end')

    dot.edge('for_end', 'q_check', label='No\\n(ç¶™ç¶š)')
    dot.edge('for_end', 'fail', label='Yes')

    return dot

# ========================================
# ãƒ¡ã‚¤ãƒ³å®Ÿè¡Œéƒ¨åˆ†
# ========================================

print("=" * 60)
print("n_queené–¢æ•° ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆç”Ÿæˆãƒ—ãƒ­ã‚°ãƒ©ãƒ ")
print("=" * 60)
print()

# ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆç”Ÿæˆ
print("ğŸ“Š n_queené–¢æ•°ã®ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆã‚’ç”Ÿæˆä¸­...")
flowchart = create_n_queen_flowchart()

# ç”»åƒã‚µã‚¤ã‚ºã¨è§£åƒåº¦ã‚’èª¿æ•´ (å¤§ãã„ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆãªã®ã§ç¸¦é•·ã«)
flowchart.attr(dpi='150')  # è§£åƒåº¦
flowchart.attr(size='10,16')  # ã‚µã‚¤ã‚º (ã‚¤ãƒ³ãƒ) - ç¸¦é•·ã«è¨­å®š

# PNGãƒ•ã‚¡ã‚¤ãƒ«ã§ä¿å­˜
output_path = flowchart.render('n_queen_flowchart', format='png', cleanup=True)
print(f"âœ… ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆãŒä¿å­˜ã•ã‚Œã¾ã—ãŸ: {output_path}")
print()

# Colabã§ç›´æ¥è¡¨ç¤º
print("=" * 60)
print("ğŸ“Œ ãƒ•ãƒ­ãƒ¼ãƒãƒ£ãƒ¼ãƒˆãƒ—ãƒ¬ãƒ“ãƒ¥ãƒ¼")
print("=" * 60)
display(flowchart)

# Google Colabã§ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰
print()
print("=" * 60)
print("ğŸ“¥ ãƒ•ã‚¡ã‚¤ãƒ«ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰")
print("=" * 60)

try:
    from google.colab import files
    print("n_queen_flowchart.png ã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ä¸­...")
    files.download('n_queen_flowchart.png')
    print()
    print("âœ… ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰å®Œäº†!")

except ImportError:
    print("âš ï¸ Google Colabç’°å¢ƒã§ã¯ã‚ã‚Šã¾ã›ã‚“ã€‚")
    print(f"ãƒ•ã‚¡ã‚¤ãƒ«ã¯ãƒ­ãƒ¼ã‚«ãƒ«ã«ä¿å­˜ã•ã‚Œã¾ã—ãŸ: {output_path}")

print()
print("=" * 60)
print("ğŸ‰ å®Œäº†!")
print("=" * 60)
```
