```python
word2freq = {}
with open('/content/drive/MyDrive/AI/jp_corpus.txt') as f:
  for data_line in f:
    data_line = data_line.rstrip()
    doc = nlp(data_line)
    for token in doc:
      #print(token.text)
      #print(word2freq[token])
      print(word2freq.get(token.text, 100))
      word2freq[token.text] = word2freq.get(token.text, 0) + 1
```

```python
for word, freq in word2freq.item():
  print(word, freq)
```
