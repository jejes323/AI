```python
import wikipedia

# 1. input()を用いて標準入力からユーザが入力する
keyword = input()

wikipedia.set_lang("ja")
search_response = wikipedia.search(keyword)

# 2. 入力した検索結果が．0 件のとき
if len(search_response) == 0:
  print('該当するページがありません')
else:
  print('キーワード:' + str(keyword) + 'での検索結果は' + str(len(search_response)) + '件です。')
  print(search_response)

# 3. 最初の 1 件のページタイトルと、本文の最初の 300 文字
  page_data = wikipedia.page(search_response[0])
  title = page_data.title
  content = page_data.content[:300]

  print("\n--- タイトル ---\n")
  print(title)
  print("--- 本文の最初の300文字 ---")
  print(content)
```
