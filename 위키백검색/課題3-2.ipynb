```python
import wikipedia
import spacy
from collections import Counter

# Wikipediaから甲南大学のページを読み込む
wikipedia.set_lang("ja")
try:
  page = wikipedia.page("甲南大学")
except wikipedia.exceptions.PageError:
    print("ページが見つかりませんでした。")
else:
  text = page.content

# spaCyで日本語解析
  nlp = spacy.load('ja_ginza')

# Increase the max_length to accommodate the longer text
  nlp.max_length = 500000 # or any value larger than 181167
  m = nlp(text)

  sentences = []
  for sent in m.sents:
    if "学生" in sent.text:
      sentences.append(sent.text)

  print("=== 学生を含む文（先頭10文） ===")

  count = 1
  for sentence in sentences:
      print(count, ')', sentence)
      count = count + 1
      if count == 11:
          break
```
