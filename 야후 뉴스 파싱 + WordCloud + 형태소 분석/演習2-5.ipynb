```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/AI/miyajima_sample.csv', encoding='utf8',parse_dates=['旅行の時期'])

#口コミの件数の算出
count=len(df)
print("口コミの件数", count)

#口コミ平均文字数の算出
word_count=0
for review in df["口コミ"]:
  word_count+=len(review)
  print("口コミ平均文字数", round(word_count/count,1))
```

```python
# スクレイピング：Web ページからデータをとってくる
# 1. BeatifulSoup の基本 <h1>の中身を出力してみよう．

from bs4 import BeautifulSoup

html = "<html><body><h1>Hello!</h1></body></html>"
soup = BeautifulSoup(html, "html.parser")
print(soup.h1.text)
```

```python
# 2. Yahoo!のニュースをとってきてみよう．以下を記述して，Yahoo!のニュースのソースをとってこよう

import requests
from bs4 import BeautifulSoup

url='http://news.yahoo.co.jp'
html = requests.get(url)
contents =BeautifulSoup(html.content,"html.parser")
print(contents)
```

```python
# 3. Yahoo!ニュースから，タイトルだけをとってこよう！ 以下を打ち込んで，Yahoo!のニュースのタイトルをとってこよう

import requests
from bs4 import BeautifulSoup
import re


url = "https://www.yahoo.co.jp/"
req = requests.get(url)
contents = BeautifulSoup(req.content, "html.parser")
data_list = contents.find_all(href=re.compile("news.yahoo.co.jp/pickup"))
text1 = ""

#data.text のタグの中に書かれた見出しに改行を入れて text1 に追加する．
for data in data_list:
  text1 += data.text + "¥n"
  print(text1)
```
