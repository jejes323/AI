メイン処理:ファイルの読み読み込みと内容の出力

```python
word_lengths = []
with open('/content/drive/MyDrive/AI/jp_corpus.txt') as f:
  for data_line in f:
    data_line = data_line.rstrip()
    words = data_line.split(' ')
    for word in words:
      word_lengths.append(len(word))

print(word_lengths)
```

```python
import matplotlib.pyplot as plt
plt.hist(word_lengths, bins=20)
plt.show()
```
