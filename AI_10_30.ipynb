```python
stock = [
    {'apple': 6, 'orange': 15, 'grape': 10},
    {'apple': 2, 'orange': 9,  'grape': 1},
    {'apple': 1, 'orange': 3,  'grape': 12}
]

points = {'Aoki':3, 'Ito': 3, 'Tanaka':3}

for i in stock:

  if i['apple'] >= 3:
    points['Aoki'] += 2
  elif i['apple'] < 3 and i['orange'] >= 8:
    points['Ito'] -= 1
  else:
    points['Tanaka'] +=5

  if i['grape'] >= 10:
    points['Aoki'] += 5
    points['Tanaka'] += 5
    points['Ito'] += 5

print(points)
```
