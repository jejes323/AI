メイン処理:ファイルの読み読み込みと内容の出力

```python
long_words = []

with open('/content/drive/MyDrive/AI/Country.txt') as f:
    for data_line in f:
        data_line = data_line.rstrip()
        words = data_line.split()
        for word in words:
            if len(word) >= 2:
                long_words.append(word)

print(long_words)
```
